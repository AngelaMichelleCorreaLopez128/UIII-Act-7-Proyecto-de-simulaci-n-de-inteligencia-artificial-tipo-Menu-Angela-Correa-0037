#### Ejercicio 14: Panadero
El programa es un simulador interactivo de una panadería diseñado para enseñar lógica básica. Funciona mediante un ciclo while que mantiene la "tienda" abierta hasta que el usuario decide salir.

```python
print("--- Panadería Inteligente ---")
cliente_en_tienda = "si"

while cliente_en_tienda == "si":
    print("\nMenú: 1. Ver pan | 2. Recomendación IA | 3. Salir")
    opcion = input("¿Qué deseas hacer? ")

    if opcion == "1":
        print("Panadero: Hoy tenemos Conchas y Bolillos.")
    elif opcion == "2":
        clima = input("¿Hace frío hoy? (si/no): ")
        if clima == "si":
            print("IA: El clima está perfecto para un chocolate y una Concha.")
        elif clima == "no":
            print("IA: Te recomiendo un Bolillo para un sándwich fresco.")
        else:
            print("IA: No entendí eso, ¡pero el pan siempre es buena idea!")
    elif opcion == "3":
        cliente_en_tienda = "no"
        print("¡Gracias por visitarnos!")
    else:
        print("Opción no válida, intenta de nuevo.")
print("Programa realizado por Angela Michelle Correa Lopez NL = 14")
```
